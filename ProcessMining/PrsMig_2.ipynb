### Анализ событий

Имеем в распоряжении журнал событий процесса, изучим входящий в него набор событий и их базовые статистики.
Создадим копию нашего исходного датасета и запишем ее в переменную events. 
Это нужно, чтобы не изменять исходные данные, если такая надобность появится.

```python
events = df.copy()
```

#### Получение базовых статистических характеристик

```python
# считаем базовые метрики:
# 1. количество появления события в данных
# 2. количество кейсов, в которых встречается событие
# 3. количество пользователей, исполнявших событие
event_stat = events.groupby('EVENT_NAME', as_index=False).agg({'CASE_ID' : ['count', 'nunique'], 'EVENT_USER' : 'nunique'})
print(event_stat)
```

Из-за того, что к одному полю применялось более одного типа агрегации, структура колонок в датафрейме приняла мультиуровневый формат, который не очень подходит для последующей визуализации результатов. 

```python
# меняем наименования столбцов
event_stat.columns = event_stat.columns.map('_'.join)
event_stat.rename(columns={'EVENT_NAME_' : 'EVENT_NAME', 'CASE_ID_count' : 'ev_cnt'}, inplace=True)
print(event_stat)
```

Теперь для каждого события мы знаем частоту его встречаемости среди событий и кейсов. Мы также можем получить данные об общем количестве кейсов и событий в нашем. 

Используя эти данные для каждого события в таблице определим:

его долю от общего числа событий
в каком % кейсов от общего числа кейсов событие встречается
Также отсортируем расширенную таблицу по частоте встречаемости события и выведем топ-10 самых популярных событий.

```python
# считаем долю события от общего числа событий
event_stat['%ev_cnt'] = (event_stat['ev_cnt'] / event_stat['ev_cnt'].sum()) * 100

# считаем, в каком % кейсов встречается событие
event_stat['%ev_cases_occasion'] = (event_stat['CASE_ID_nunique'] / events['CASE_ID'].nunique()) * 100

event_stat = event_stat.sort_values('ev_cnt', ascending = False).reset_index(drop=True)

event_stat.head(10)
```

Воспользуемся библиотекой seaborn. На одной оси будем показывать численность событий, а на другой - %кейсов, в которых это событие встречается.

```python
# визуализируем полученную таблицу
ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots(figsize=(12,6));
fig.autofmt_xdate(rotation=90);

sns.barplot(data = event_stat, x='EVENT_NAME', y='ev_cnt', alpha=0.5, ax=ax1);
ax2 = ax1.twinx()

sns.lineplot(data = event_stat['%ev_cases_occasion'], marker='o', sort = False, ax=ax2);

plt.title('Статистика событий: Численность общая и встречаемость по кейсам');
```

#### Выводы по статистическим характеристикам:

Топ-5 самых популярных событий в имеющихся данных: F, A, D, E, H
Суммарно это события составляют около 60% всех событий.

Топ-10 самых часто встречающихся событий в кейсах: F, A, H, J, M, C, B, N, O

Интересно заметить, что в данных имеются события, которые встречаются довольно часто и занимают большую долю от общего числа событий, но при этом происходят не во всех кейсах. Например, события D, E, L занимают 3, 4 и 7 место среди доли от общего числа событий, но при этом событие L встречается лишь в 50% кейсов. Это означает, что у данных событий имелось большое количество повторений в рамках каких-то конкретных кейсов, что может быть целью для дальнейшего анализа.

### Анализ переходов

Частотные статистики по событиям, имеющимся в наших данных посчитаны. Для извлечения больших знаний полезно узнать о длительностях этих событий.
Переходом в рамках какого-то конкретного экземпляра процесса является любая ненулевая последовательность событий.

```python
# Создадим копию датафрейма events из предыдущего шага и запишем ее в переменную transitions.
transitions = events.copy()
transitions.head()
```

в этом датафрейме нет переходов в виде отдельного поля. Значит, давайте добавим их самостоятельно. Добавим следующие столбцы:

```python
# следующее событие
transitions['NEXT_EVENT_NAME'] = transitions.groupby(['CASE_ID'])['EVENT_NAME'].shift(-1)

# дата следующего события
transitions['NEXT_EVENT_TIME'] = transitions.groupby(['CASE_ID'])['EVENT_TIME'].shift(-1)

# длительность перехода
transitions['TRANSITION_DUR_SEC'] = (transitions['NEXT_EVENT_TIME'] - transitions['EVENT_TIME']).apply(lambda x: x.total_seconds())
transitions['TRANSITION_DUR_MIN'] = transitions['TRANSITION_DUR_SEC'] / 60
transitions['TRANSITION_DUR_HRS'] = transitions['TRANSITION_DUR_MIN'] / 60
transitions['TRANSITION_DUR_DAYS'] = transitions['TRANSITION_DUR_HRS'] / 24
transitions.head()
```

Если какое-то событие было последним в рамках кейса, то для него в новых полях будут находиться пустые значения. Для таких случаев проведем следующую обработку:

NEXT_EVENT_NAME будем присваивать значение "PROCESS_END", указывая на конец кейса
Длительность переходов будем приравнивать нулю

```python
transitions.loc[transitions[pd.isnull(transitions['NEXT_EVENT_NAME']) == True].index, 'NEXT_EVENT_NAME'] = 'PROCESS_END'
transitions.loc[transitions[transitions['NEXT_EVENT_NAME'] == 'PROCESS_END'].index, 'TRANSITION_DUR_SEC'] = 0
transitions.loc[transitions[transitions['NEXT_EVENT_NAME'] == 'PROCESS_END'].index, 'TRANSITION_DUR_MIN'] = 0
transitions.loc[transitions[transitions['NEXT_EVENT_NAME'] == 'PROCESS_END'].index, 'TRANSITION_DUR_HRS'] = 0
transitions.loc[transitions[transitions['NEXT_EVENT_NAME'] == 'PROCESS_END'].index, 'TRANSITION_DUR_DAYS'] = 0
```

На базе преобразованных данных сформируем новое поле c именем перехода. Это будет простая конкатенация значений столбцов EVENT_NAME и NEXT_EVENT_NAME.

```python
# добавляем наименование перехода
transitions['TRANSITION'] = transitions['EVENT_NAME'] + ' -> ' + transitions['NEXT_EVENT_NAME']
transitions.head()
```

Данные подготовлены, значит можно считать статистики и проводить анализ.

Переходы, например, можно отобразить в виде heatmap диаграммы, построенной на базе сводной таблицы. 

```python
# считаем долю каждого перехода от общего числа перехода в датасете (в процентах)
tran_pivot = pd.crosstab(transitions['EVENT_NAME'], transitions['NEXT_EVENT_NAME']) 
tran_pivot_norm = round((tran_pivot / tran_pivot.sum().sum()) * 100, 2)

# визуализируем сводную таблицу в виде heatmap
plt.figure(figsize=(30, 15))
sns.heatmap(tran_pivot_norm, annot = True);
```

Heatmap диаграммы переходов бывают очень полезны, когда требуется посмотреть, существуют ли в процессе связанные "группы" переходов. Как можем заметить на этом примере, таких групп нет. Имеются лишь отдельные виды переходов, которые встречаются достаточно часто. Например, A-B, B-C, F-H занимает наибольшую долю + мы получили информацию, что после события D чаще всего следует окончание кейса.

Получим информацию о длительностях переходов в группировках

```python
# считаем статистику по переходам:
# количество переходов
# количество кейсов, в которых встречается переход
# средняя и медианная длительность перехода в минутах
transition_stat = transitions.groupby('TRANSITION', as_index=False).agg({'EVENT_NAME' : 'count',
                                                                    'CASE_ID' : 'nunique',
                                                                    'TRANSITION_DUR_MIN' : ['mean', 'median']})
transition_stat.columns = transition_stat.columns.map('_'.join)
transition_stat.rename(columns={'TRANSITION_' : 'TRANSITION',
                      'EVENT_NAME_count' : 'tr_count',
                     }, inplace=True)
```

Добавим столбцы с долями от общего числа переходов и кейсов + отсортируем новый датафрейм.

```python
# считаем долю перехода от общего количества переходов
transition_stat['%tr_count'] = (transition_stat['tr_count'] / transition_stat['tr_count'].sum()) * 100

# считаем, в каком % кейсов встречается переход
transition_stat['%cases'] = (transition_stat['CASE_ID_nunique'] / transitions['CASE_ID'].nunique()) * 100

transition_stat = transition_stat.sort_values(['%tr_count', 'TRANSITION_DUR_MIN_median'], ascending=False).reset_index(drop=True)
```

```python
transition_stat.head(5)
```

Практически во всех топ-10 переходах средние и медианные длительности сильно отличаются, это свидетельствует о наличии в процессе аномально длинных переходов, которые смещают среднее значение длительности вправо. Также в процессе есть переходы с достаточно долгой медианной длительностью. Такие переходы при большой массовости (например M -> N) вполне могут выступать "узкими горлышками" процесса и потенциальными кандидатами на проведение оптимизационных мероприятий.

По аналогии с прошлым шагом, можем в seaborn построить визуализацию статистики по топ-10 переходам.

```python
ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots(figsize=(12,6))
fig.autofmt_xdate(rotation=90);

sns.barplot(data = transition_stat.head(10), x='TRANSITION', y='tr_count', alpha=0.5, ax=ax1);
ax2 = ax1.twinx()

sns.lineplot(data = transition_stat['%cases'].head(10), marker='o', sort = False, ax=ax2);

plt.title('Статистика переходов: Численность общая и встречаемость по кейсам');
```

### Анализ кейсов

Изучив данные событий и переходов, давайте перейдем на уровень анализа кейсов и вариантов.

Кейс - конкретный экземпляр бизнес-процесса, который содержится в наших данных. Посчитав статистики в детализации по кейсам, мы сможем понять, как ведет себя процесс в целом, и найти конкретные болевые точки, которые потребуют дополнительного изучения.

При анализе кейсов нам может быть полезно найти ответы на следующие вопросы:

Как распределяются экземпляры процессов по количеству уникальных / неуникальных событий, длительности, количеству участников?
К какому варианту исполнения относится тот или иной кейс?
Какие имеются взаимосвязи между характеристиками процесса? Как те или иные характеристики оказывают влияние на его длительность?

#### Подсчет основных статистик кейсов

```python
# считаем базовые агрегаты по кейсам
# 1. количество событий
# 2. количество уникальных событий
# 3. последовательность событий (вариант исполнения)
# 4. минимальную и максимальную дату события
# 5. количество уникальных пользователей
cases = transitions.groupby('CASE_ID').agg({'EVENT_NAME' : ['count', 'nunique', ' -> '.join],
                                   'EVENT_TIME' : ['min', 'max'],
                                   'EVENT_USER' : 'nunique'}).reset_index()

cases.columns = cases.columns.map('_'.join)
cases.rename(columns={'CASE_ID_' : 'CASE_ID',
                      'EVENT_NAME_join' : 'variant'
                     }, inplace=True)
```

```python
print(cases)
    
```

Посчитаем события, которые исполнялись наименьшим количеством участников.

```python
min_user = transitions.groupby('EVENT_NAME').agg({'EVENT_USER' : 'count'}).reset_index()

min_user = min_user.sort_values(['EVENT_USER'], ascending=False).reset_index(drop=True)

min_user = min_user.sort_values('EVENT_USER', ascending = True).reset_index(drop=True)
print(min_user)
```

Посчитаем медианную длительность кейсов

```python
case_min = transitions.groupby('CASE_ID').agg({'EVENT_TIME' : 'min'}).reset_index()
case_max = transitions.groupby('CASE_ID').agg({'EVENT_TIME' : 'max'}).reset_index()
case_dif = case_max['EVENT_TIME'] - case_min['EVENT_TIME']
dif = statistics.median(case_dif)
print(dif)
```

Посчитаем дополнительные показатели

доля уникальных событий от общего числа событий (внутри кейса)
длительность каждого кейса (конец - начало) в минутах

```python
# считаем дополнительные показатели
# час, когда было выполнено первое действие по кейсу
cases['first_event_hr'] = cases['EVENT_TIME_min'].dt.hour

# доля уникальных событий от общего числа событий (внутри кейса)
cases['%unique'] = cases['EVENT_NAME_nunique'] / cases['EVENT_NAME_count']

# длительность каждого кейса (конец - начаоло)
cases['dur'] = cases['EVENT_TIME_max'] - cases['EVENT_TIME_min']

# длительность каждого кейса в минутах и часах
cases['dur_mins'] = cases['dur'].apply(lambda x: x.total_seconds() / 60)
cases['dur_hrs'] = cases['dur_mins'] / 60
cases['dur_days'] = cases['dur_hrs'] / 24

# # после перевода в минуты, часы и дни удаляем поле dur типа timestamp. оно нам не потребуется
del cases['dur']
```

### Поиск взаимосвязей между переменными

Чтобы понять, как взаимосвязаны друг с другом переменные процессов, посчитаем вхождение каждого типа события в каждый кейс.

```python
# считаем количество вхождений каждого события в каждый кейс
event_agg = pd.crosstab(transitions['CASE_ID'], transitions['EVENT_NAME']).reset_index()
event_agg.columns.name = None
```

Расширим датафрейм cases полученными данным о событиях внутри каждого кейса, соединив cases и event_agg при помощи метода merge().

```python
# собираем все в единую таблицу кейсов
cases_full = pd.merge(cases, event_agg, 'left', on=['CASE_ID'])
cases_full.head(2)
```

Помочь найти взаимосвязь между переменными поможет корреляция и метод corr(). Корреляция как раз показывает, существуют ли в имеющемся наборе данных линейные связи между числовыми величинами. Чтобы было проще интерпретировать получаемую корреляционную матрицу, представим ее в виде heatmap диаграммы.

```python
# визуализируем корреляционную матрицу
plt.figure(figsize=(60, 30))
sns.heatmap(cases_full.corr(), annot = True);
```

Чтобы определить, какие переменные оказывают негативное влияние на длительность процесса, посчитаем корреляцию только с переменной dur_days, используя метод corrwith().

```python
# считаем корреляцию только с целевой переменной - длительностью
dur_corr = pd.DataFrame(cases_full.corrwith(cases_full['dur_days'])).reset_index()
dur_corr.columns = ['col', 'corr']
dur_corr = dur_corr.sort_values('corr')
```

```python
# визуализируем корреляции в виде столбчатой диаграммы
ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots(figsize=(12,6))
fig.autofmt_xdate(rotation=90);

plt.title('Корреляция атрибутов кейсов с длительностью');

graph = sns.barplot(data=dur_corr, x='col', y='corr');
graph.axhline(0.4, color='red', linestyle='--')
graph.axhline(-0.4, color='red', linestyle='--')

plt.show();
```

Выводы из графика корреляции.

Имеется достаточно сильная обратная линейная связь между %unique и длительностью процесса. Уменьшение доли уникальных событий от общего числа событий приводит к увеличению длительности. Другими словами, наличие петель в процессах увеличивают длительность последних.
Чем больше внутри процесса событий L и I, тем длиннее исполнение процесса. 
Выводы выглядят достаточно логичными. Повторяющиеся события, передача ответственности за инцидент от одной рабочей группы к другой действительно могут пагубно влиять на длительность исполнения. Значит, нужно детальнее анализировать процессы с большим количеством смен рабочих групп, исследовать причины и контекст возникновения таких смен, чтобы не допускать их в будущем.

### Анализ вариантов процесса

Анализ вариантов является одним из самых распространенных методов, применяемых в Process Mining, потому что помогает понять:

Какие сценарии исполнения имеются в рассматриваемом процессе? Насколько часто они происходят?
Насколько вариативным является процесс? Если вариантов исполнения слишком много, это может говорить о неструктурированности процесса.
Какие сценарии процесса являются успешными и могут быть взяты как "best practice" для тиражирования? От каких вариантов лучше отказаться?

Например, если мы имеем кейсы:

A - B - C - D
A - C - D - B
A - B - C - D
То общее количество кейсов у нас будет 3, а вариантов исполнения кейсов будет 2, потому что по своему содержанию кейсы 1 и 3 являются идентичными.

В группировке по вариантам посчитаем:

количество кейсов
среднее и медианное количество пользователей
среднюю и медианную длительность
количество событий

```python
# считаем статистики по вариантам:
# количество кейсов
# среднее и медианное количество пользователей
# среднюю и медианную длительность
variant_stat = cases.groupby('variant', as_index=False).agg({
                              'CASE_ID' : 'count',
                              'EVENT_NAME_count' :  ['mean'],
                              'EVENT_USER_nunique' : ['mean', 'median'],
                              'dur_mins' : ['mean', 'median'],
                              'dur_hrs' : ['mean', 'median'],
                              'dur_days' : ['mean', 'median']})

variant_stat.columns = variant_stat.columns.map('_'.join)
variant_stat.rename(columns={'variant_' : 'variant'}, inplace=True)
```

Дополнительно для каждого варианта посчитаем долю от общего числа кейсов в наших данных.

```python
# для каждого варианта считаем, какой % кейсов покрывает
variant_stat['%cases'] = (variant_stat['CASE_ID_count'] / variant_stat['CASE_ID_count'].sum()) * 100
```

Провизуализируем топ-10 вариантов в абсолютных и относительных величинах.

```python
ax1 = sns.set_style(style=None, rc=None )

fig, ax1 = plt.subplots(figsize=(12,6))
fig.autofmt_xdate(rotation=45);

sns.barplot(data = variant_stat.sort_values('%cases', ascending=False).reset_index(drop=True).head(10), x='variant', y='CASE_ID_count', alpha=0.5, ax=ax1);
ax2 = ax1.twinx()

sns.lineplot(data = variant_stat.sort_values('%cases', ascending=False).reset_index(drop=True)['%cases'].head(10), marker='o', sort = False, ax=ax2);

plt.title('Статистика вариантов: Численность общая и встречаемость в процессе');
```

Видим, что каждый из топ-10 вариантов покрывает менее 3,5% от общего числа кейсов. При этом первые 4 варианта являются наиболее распространенными, суммарно покрывая около 14% всех исполнений процессов.
